```python
# Program prediksi kelulusan mahasiswa
# Input berupa IPS semester 1 sampai 8
# Output berupa kategori kelulusan A, B, C, atau D
# Model yang digunakan adalah Decision Tree

# Import library yang diperlukan
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# Membaca dataset dari file CSV
df = pd.read_csv("df.csv")

# Menampilkan informasi awal dataset
print("Dataset dimuat")
print(df.head())

# Menentukan kolom fitur berupa IPS semester 1 sampai 8
fitur = ["ips1", "ips2", "ips3", "ips4", "ips5", "ips6", "ips7", "ips8"]

# Menentukan kolom target berupa label kelulusan
target = "label"

# Memisahkan data fitur dan data target
X = df[fitur]
y = df[target]

# Membuat encoder untuk mengubah label huruf menjadi angka
le = LabelEncoder()

# Melakukan encoding label A, B, C, dan D
y_encoded = le.fit_transform(y)

# Menampilkan hasil mapping label ke angka
print("Mapping Label:", dict(zip(le.classes_, le.transform(le.classes_))))

# Membagi data menjadi data latih dan data uji
# 80 persen untuk training dan 20 persen untuk testing
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42
)

# Membuat model Decision Tree
# Menggunakan entropy sebagai kriteria pemisahan
# Kedalaman pohon dibatasi agar tidak overfitting
model = DecisionTreeClassifier(
    criterion="entropy",
    max_depth=6,
    random_state=42
)

# Melatih model menggunakan data training
model.fit(X_train, y_train)
print("Model selesai dilatih")

# Melakukan prediksi pada data testing
y_pred = model.predict(X_test)

# Menampilkan nilai akurasi model
print("Akurasi:", accuracy_score(y_test, y_pred))

# Menampilkan laporan evaluasi klasifikasi
print("\nLaporan:\n", classification_report(
    y_test,
    y_pred,
    target_names=le.classes_
))

# Meminta input IPS mahasiswa secara manual
print("\nINPUT DATA MAHASISWA")

# List untuk menyimpan nilai IPS
ips = []

# Mengambil input IPS semester 1 sampai 8 dari pengguna
for i in range(1, 9):
    nilai = float(input(f"IPS Semester {i}: "))
    ips.append(nilai)

# Mengubah data input menjadi DataFrame agar sesuai dengan model
data_input = pd.DataFrame([ips], columns=fitur)

# Melakukan prediksi kelulusan berdasarkan input pengguna
hasil = model.predict(data_input)

# Mengubah hasil prediksi angka menjadi label huruf
hasil_label = le.inverse_transform(hasil)[0]

# Menampilkan hasil prediksi kelulusan
print("\nHASIL PREDIKSI")
print("Hasil prediksi Kelulusan Anda adalah ", end='')

if hasil_label == 'A':
    print("Lulus Dengan Sangat Baik")
elif hasil_label == 'B':
    print("Lulus Dengan Baik")
elif hasil_label == 'C':
    print("Lulus Dengan Cukup")
elif hasil_label == 'D':
    print("Tidak Lulus")
else:
    print("Tidak Lulus")
```
